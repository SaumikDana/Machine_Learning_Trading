```python
import pandas as pd  # pip install pandas
import requests  # pip install requests
from bs4 import BeautifulSoup  # pip install beautifulsoup4
import datetime
import matplotlib.pyplot as plt  # pip install matplotlib

# Scrape data from MarketWatch
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:87.0) Gecko/20100101 Firefox/87.0',
}
url = 'https://www.marketwatch.com/tools/earningscalendar'
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

tables = soup.find_all('table')
current_datetime = datetime.datetime.now().strftime('%m-%d-%y %H_%M_%S')
xlsxwriter = pd.ExcelWriter(f'Earning Calendar ({current_datetime}).xlsx', engine='xlsxwriter')

for i, table in enumerate(tables):
    if not table.find('tbody').find_all('tr'):
        continue

    earning_date = f'Earnings_{i+1}'
    print(earning_date)
    df = pd.read_html(str(table))[0]

    # Extract and print ticker symbols if the column exists
    if 'Symbol' in df.columns:
        ticker_symbols = df['Symbol'].dropna().unique()
        print(f"Ticker Symbols in {earning_date}: {ticker_symbols}")

    df.to_excel(xlsxwriter, sheet_name=earning_date, index=False)

xlsxwriter.close()
print('Earning tables Excel file exported')

# Read and analyze the Excel file
file_name = f'Earning Calendar ({current_datetime}).xlsx'
all_sheets = pd.read_excel(file_name, sheet_name=None)

for sheet_name, data in all_sheets.items():
    print(f"Analyzing {sheet_name}...")

    # Display ticker symbols if the column exists
    if 'Symbol' in data.columns:
        ticker_symbols = data['Symbol'].dropna().unique()
        print(f"Ticker Symbols in {sheet_name}: {ticker_symbols}")

    summary = data.describe()
    print(summary)

    if 'EPS Forecast' in data.columns and pd.api.types.is_numeric_dtype(data['EPS Forecast']):
        plt.figure()
        data['EPS Forecast'].hist(bins=20)
        plt.title(f'EPS Forecast Histogram - {sheet_name}')
        plt.xlabel('EPS Forecast')
        plt.ylabel('Frequency')
        plt.show()
```
