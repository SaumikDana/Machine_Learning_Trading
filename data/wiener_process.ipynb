```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_steps = 1000          # Number of steps
delta_t = 1/365           # Time step size (e.g., in days)
sqrt_delta_t = np.sqrt(delta_t)
sigma = 0.1               # Volatility parameter

# Initialize arrays to store steps and position
time = np.linspace(0, num_steps*delta_t, num_steps)
wiener_process = np.zeros(num_steps)

# Generate the Wiener process - cumulative sum of normal random variables
for i in range(1, num_steps):
    # Random increment: normal random variable with mean 0 and variance delta_t
    wiener_process[i] = wiener_process[i-1] + np.random.normal(0, sigma * sqrt_delta_t)

# Plotting the Wiener process
plt.figure(figsize=(10, 6))
plt.plot(time, wiener_process, label='Wiener Process')
plt.title('Wiener Process')
plt.xlabel('Time')
plt.ylabel('W(t)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_steps = 1000          # Number of steps
S0 = 100                  # Initial stock price
T = 1                     # Time horizon in years
mu = 0.05                 # Expected return
sigma = 0.2               # Volatility
dt = T / num_steps        # Time step size
t = np.linspace(0, T, num_steps)  # Time vector

# Simulate the random part of the Geometric Brownian Motion
W = np.random.standard_normal(size=num_steps)
W = np.cumsum(W)*np.sqrt(dt) # Standard Wiener Process

# Calculate the stock price path
S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)

# Plotting the stock price path
plt.figure(figsize=(10, 6))
plt.plot(t, S, label='Stock Price')
plt.title('Stock Price Simulation using Geometric Brownian Motion')
plt.xlabel('Time in years')
plt.ylabel('Stock Price S(t)')
plt.legend()
plt.show()
```
