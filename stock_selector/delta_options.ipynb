```python
import matplotlib.pyplot as plt
import numpy as np

# Delta ranges for each option type
delta_ranges = {
    'OTM Calls': (0.01, 0.49),
    'ATM Calls': (0.5, 0.5),
    'ITM Calls': (0.51, 0.99),
    'OTM Puts': (-0.49, -0.01),
    'ATM Puts': (-0.5, -0.5),
    'ITM Puts': (-0.99, -0.51)
}

# Plot settings
fig, ax = plt.subplots(figsize=(10, 6))
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan']
option_types = list(delta_ranges.keys())
markers = ['o', 's', 'D', 'o', 's', 'D']

# Plotting each option type delta range
for i, (option, deltas) in enumerate(delta_ranges.items()):
    if "ATM" in option:  # ATM options have a single delta value
        ax.plot([i + 1], [deltas[0]], color=colors[i], marker=markers[i], markersize=10, label=option)
    else:  # OTM and ITM options have a range of delta values
        ax.plot([i + 1, i + 1], [deltas[0], deltas[1]], color=colors[i], marker=markers[i], label=option)

# Customizing the plot
ax.set_xticks(range(1, len(option_types) + 1))
ax.set_xticklabels(option_types, rotation=45, ha='right')
ax.set_ylabel('Delta Value')
ax.set_title('Delta Ranges for Different Option Types')
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.axhline(0, color='black', linewidth=1)  # Add a line at Delta = 0
ax.legend()

plt.tight_layout()
plt.show()
```
