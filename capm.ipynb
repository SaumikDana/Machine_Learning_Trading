```python
import matplotlib.pyplot as plt

# Data for CAPM
beta = [0.5, 1.0, 1.5, 2.0]  # Example betas for different assets
risk_free_rate = 0.05  # Risk-free rate (5%)
market_premium = 0.08  # Market premium (8%)
expected_return = [risk_free_rate + b * market_premium for b in beta]  # Calculate expected return

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(beta, expected_return, marker='o')

# Adding titles and labels
plt.title('Capital Asset Pricing Model (CAPM)')
plt.xlabel('Beta (Systematic Risk)')
plt.ylabel('Expected Return')
plt.grid(True)

# Annotating risk-free rate
plt.axhline(y=risk_free_rate, color='r', linestyle='--')
plt.text(2.1, risk_free_rate, 'Risk-Free Rate', verticalalignment='center', color='red')

# Annotating the market rate
market_return = risk_free_rate + 1 * market_premium  # Market rate at beta=1
plt.axhline(y=market_return, color='g', linestyle='--')
plt.text(2.1, market_return, 'Market Return', verticalalignment='center', color='green')

# Show plot
plt.show()
```
